# 3. Process a 3-channel colour image

Using matplotlib, read the colour image stored in file `stickanimals.jpg` and perform the following tasks:

1. Write a function `colour_rowscolumns(image)` that determines the number of rows and columns in the image.
2. Write a function `remove_red` that has the red channel removed. You can achieve this by setting all values of the red channel to zero.
4. Write a function `to_greyscale` that creates a new image that is the greyscale version of the original image. To convert the image to greyscale, each pixel in the new image should be computed as the average of the channel values of the corresponding pixel in the original image.

Below are examples of use of these functions.

```python
from matplotlib import pyplot as plt
colourimage = plt.imread('stickanimals.jpg')
import week1
plt.imshow(colourimage)
plt.axis('off')
```

```python
week1.colour_rowscolumns(colourimage)
```

```python
nored = week1.remove_red(colourimage)
plt.imshow(nored)
plt.axis('off')
```

```python
greyversion = week1.to_greyscale(colourimage)
plt.imshow(greyversion, cmap='gray')
plt.axis('off')
```
